```python
preprocessed_data=pd.merge(x,y,left_index=True,right_index=True)

preprocessed_data
```

```python
y.value_counts().plot(kind='pie', autopct='%1.0f%%')
```

```python
x=preprocessed_data.iloc[:,:-1]
y=preprocessed_data.iloc[:,-1]

x,y
```

```python
y = y.replace([5, 4, 3, 2, 1], 0)
y = y.replace([7, 6], 1)
```

```python
y.describe()
```

```python
#SPLIT THE DATASET

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
type(x_train)
```

```python
#FEATURE SCALING

from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler

mm_scale=preprocessing.MinMaxScaler(feature_range=(0,1))

x_train=mm_scale.fit_transform(x_train)
x_test=mm_scale.transform(x_test)


len(x_train[0]),len(x_train)
```

```python
#PCA


from sklearn.decomposition import PCA

pca1=PCA(0.95)
x_pca=pca1.fit_transform(x_train)
x_pca_test=pca1.transform(x_test)

pca1.n_components_
```

```python
print(pca1.explained_variance_ratio_.sum())
print(pca1.components_.shape)
print(pca1.explained_variance_ratio_)
```

```python
x_pca
```

```python
from sklearn.linear_model import LogisticRegression
classifier1 = LogisticRegression(random_state=0,solver="saga",C=0.01)#,max_iter=1000)
classifier1.fit(x_pca,y_train)

y_pred1 = classifier1.predict(x_pca_test)


from sklearn.metrics import accuracy_score,mean_squared_error
print(accuracy_score(y_test,y_pred1))
print("\n")
print(mean_squared_error(y_test,y_pred1))
print(mean_squared_error(y_train,classifier1.predict(x_pca)))

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred1)
print(cm)



from sklearn.model_selection import cross_val_score

cross_val_score(classifier1,x_pca,y_train,cv=3,scoring="accuracy")
```

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(max_iter=20,tol=np.infty)
sgd_clf.fit(x_pca,y_train)


y_pred = sgd_clf.predict(x_pca_test)


from sklearn.metrics import accuracy_score,mean_squared_error
print(accuracy_score(y_test,y_pred))
print("\n")
print(mean_squared_error(y_test,y_pred))
print(mean_squared_error(y_train,sgd_clf.predict(x_pca)))

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn import tree

classifier2  = tree.DecisionTreeClassifier(max_depth=3,max_leaf_nodes=2)
classifier2.fit(x_pca,y_train)

predictions = classifier2.predict(x_pca_test)

from sklearn.metrics import accuracy_score,f1_score,mean_squared_error
print(accuracy_score(y_test,predictions))
print(f1_score(y_test,predictions, average = 'micro'))
print("\n")
print(mean_squared_error(y_test,predictions))
print(mean_squared_error(y_train,classifier2.predict(x_pca)))

from sklearn.metrics import confusion_matrix


cm=confusion_matrix(y_test,predictions)
cm
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=3, random_state=0,max_leaf_nodes=2)

clf.fit(x_pca, y_train)

predictions = clf.predict(x_pca_test)

from sklearn.metrics import accuracy_score,mean_squared_error
print(accuracy_score(y_test,predictions))
print("\n")
print(mean_squared_error(y_test,predictions))
print(mean_squared_error(y_train,clf.predict(x_pca)))


from sklearn.metrics import confusion_matrix


cm=confusion_matrix(y_test,predictions)
cm
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bag_clf = BaggingClassifier(DecisionTreeClassifier(),n_estimators=5,n_jobs=-1,max_samples=100,bootstrap=True)

bag_clf.fit(x_pca,y_train)

y_pred1 = bag_clf.predict(x_pca_test)


from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred1))

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred1)
print(cm)



from sklearn.model_selection import cross_val_score

cross_val_score(bag_clf,x_pca,y_train,cv=3,scoring="accuracy")
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

ada_clf = AdaBoostClassifier(DecisionTreeClassifier(max_depth=1),n_estimators=200,algorithm="SAMME.R",learning_rate=0.5)

ada_clf.fit(x_pca,y_train)

y_pred1 = ada_clf.predict(x_pca_test)


from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred1))

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred1)
print(cm)



from sklearn.model_selection import cross_val_score

cross_val_score(ada_clf,x_pca,y_train,cv=3,scoring="accuracy")
```
